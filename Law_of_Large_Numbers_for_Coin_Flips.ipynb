```python
import matplotlib.pyplot as plt
import random
```

```python
# 设置抛硬币的次数
num_flips = 100
# 初始化正面朝上的次数和频率列表
heads_counts = 0
frequency = []
# 抛硬币
for i in range(num_flips):
    if random.choice(('H', 'T')) == 'H':
        heads_counts += 1
    frequency.append(heads_counts / (i + 1))

# 绘制折线图
plt.plot(range(1, num_flips + 1), frequency)
plt.xlabel('Number of flips')
plt.ylabel('Frequency of heads')
plt.title('Law of Large Numbers for Coin Flips')
plt.axhline(0.5, color='r', linestyle='--') # 添加水平线 y=0.5
plt.show()
```
